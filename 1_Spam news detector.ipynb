```python
import pandas as pd
from sklearn.datasets import fetch_openml
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
```

```python
false = pd.read_csv('fake.csv')
true = pd.read_csv('real.csv')
false['label'] = 1
true['label'] = 0
data = pd.concat([false,true], ignore_index=True)
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
#check for null values
data.isnull().sum()
```

```python
lemma = WordNetLemmatizer()
stopwords = stopwords.words('english')
```

```python
nltk.download('wordnet')
```

```python
def cleaning(text):
    text = text.lower()
    text = re.sub('[^a-zA-Z]', ' ', text)
    
    token = text.split()
    
    news = [lemma.lemmatize(word) for word in token if not word in stopwords]
    cl_news =' '.join(news)
    return cl_news
```

```python
data['News'] = data['News'].apply(lambda x : cleaning(x))
```

```python
vectorizer = TfidfVectorizer(max_features = 50000, lowercase = False,ngram_range = (1,2))
```

```python
vec_news = vectorizer.fit_transform(data['News'])
news_data = pd.DataFrame(vec_news.toarray(),columns = vectorizer.get_feature_names_out())
```

```python
train_news,test_news,train_lebal,test_lebal = train_test_split(news_data,data['label'],test_size = 0.2,random_state = 0)
```
